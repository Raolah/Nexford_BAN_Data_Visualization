```python
#statistical analysis
import pandas as pd

# Example dataset with string data
data = {'type': ['Movie', 'TV Show']}
df = pd.DataFrame(data)

# Frequency count of each category
frequency = df['type'].value_counts()
print(frequency)
```

```python
DATA VISUALIZATION
```

```python
#Import matpltolib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create visualization for most watched genres which is reprsented as "type" column name in the dataset
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='type', order=df['type'].value_counts().index)
plt.title('Most Watched Genres')
plt.xticks(rotation=45)
plt.show()
```

```python
#Create visualization for ratings distribution using the "rating" column name in the dataset
plt.figure(figsize=(10,6))
sns.histplot(df['rating'], kde=True)
plt.title('Ratings Distribution')
plt.show()
```
